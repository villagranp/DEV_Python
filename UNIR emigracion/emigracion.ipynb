```python
import pandas as pd

df = pd.read_csv('emigracionCSV.csv')
dic= {'Enero': 1 ,
      'Febrero': 2 ,
      'Marzo': 3 ,
      'Abril': 4 ,
      'Mayo': 5 ,
      'Junio': 6 ,
      'Julio': 7 ,
      'Agosto': 8 ,
      'Septiembre': 9 ,
      'Octubre': 10 ,
      'Noviembre': 11 ,
      'Diciembre': 12 }
df["Mes"] = df["Mes"].replace(dic)
dftotal = df.groupby(['Mes','Año'],as_index = False).sum().pivot('Mes','Año').fillna(0)
```

```python
dfhombres = df[df.Sexo == 'Hombre']
dfmujeres = df[df.Sexo == 'Mujer']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

x_axis_labels = ['2009','2010','2011','2012']
y_axis_labels = ['Enero','Febrero','Marzo','Abril',
                 'Mayo','Junio','Julio','Agosto',
                 'Septiembre','Octubre','Noviembre','Diciembre']
plt.figure(figsize = (10,7))
plt.title(label = 'Emigración Anual 2009 - 2012', fontstyle='italic', fontsize='20' )
sns.heatmap(dftotal, cmap = "coolwarm", xticklabels=x_axis_labels, yticklabels=y_axis_labels )
plt.xlabel('Año')
plt.ylabel('Mes')

```

```python
import numpy as np
import matplotlib.pyplot as plt

barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

grupos = ['Enero','Febrero','Marzo','Abril',
                 'Mayo','Junio','Julio','Agosto',
                 'Septiembre','Octubre','Noviembre','Diciembre']
a2009 = dfhombres[dfhombres.Año == 2009]['Número emigrantes']
a2010 = dfhombres[dfhombres.Año == 2010]['Número emigrantes']
a2011 = dfhombres[dfhombres.Año == 2011]['Número emigrantes']
a2012 = dfhombres[dfhombres.Año == 2012]['Número emigrantes']
a2012[42] = 0
a2012[43] = 0
a2012[44] = 0
a2012[45] = 0
a2012[46] = 0
a2012[47] = 0
        
br1 = np.arange(len(a2009))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
 
plt.bar(br1, a2009, width=0.20, label='2009')
plt.bar(br2, a2010, width=0.20, label='2010')
plt.bar(br3, a2011, width=0.20, label='2011')
plt.bar(br4, a2012, width=0.20, label='2012')
 
plt.xticks(br1, grupos)
plt.ylabel("Personas Emigrantes", fontweight ='bold', fontsize = 15)
plt.xlabel("Año",fontweight ='bold', fontsize = 15)
plt.title('Hombres Emigrantes',fontweight ='bold', fontsize = 15)
plt.legend()
 
plt.show()
```
